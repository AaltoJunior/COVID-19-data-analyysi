Kuten huomaat kuvasta, data sisältää myös kaikki nollarivit (vasemmalla kuvaajassa vaakasuorat viivat). Päivämääriä on  niin paljon, ettei niitä voida lukea kuvaajasta. Lähdetään siis kehittämään visualisointia!

Huomataan, että ensimmäinen tautitapaus Ruotsissa on havaittu 27.2.2020. Valitaan siis analyysiin vain tuosta seuraavat päivät eli 27.2.2020 eteenpäin. Tehdään sama suomelle. Käytetään Suomenkin kohdalla Ruotsin päivämäärää, jotta datat ovat saman mittaisia. 

Pandas-kirjaston DataFrame-rakenteella on oma tyyppi datetime (Timestamp), jota se osaa käsitellä taitavasti päivämääränä. Kun päivämäärät saadaan datetime-tyyppiin, kuvaaja asettaa datapisteet automaattisesti vanhimmasta uusimpaan. Samoin se ei yritä kirjoittaa kaikkia päivämääriä kuvaajan aika-akselille(x-akselille) vaan merkkaa ainoastaan muutaman päivämäärän x-akselille. Tämä datetime-tyyppi on muotoa vuosi-kuukausi-päivä. Pandas-kirjastossa on  komento muuttamaan toista tyyppiä olevat päivämäärät sille ominaiseksi. Siirrytään nyt siis muuttamaan päivämäärän muotoa. Sen jälkeen voidaan valita analyysiin vain juuri selvitetystä ensimmäisestä havaintopäivämäärästä seuraavat päivät. Tehdään se supistamalla data sisältämään valikoidut päivämäärät.

```python
#Vaihdetaan datan päivämäärät vastaamaan Timestamp-muotoa komennolla to_datetime ja annetaan komennolle aiempi päivämäärämuoto syötteessä format.
data_ruotsi.loc[:,'Date_reported'] = pandas.to_datetime(data_ruotsi['Date_reported']) 
#Sijoitetaan datan päivämääräsarakkeeseen uutta muotoa olevat päivämäärät.
#Tulostetaan data
data_ruotsi
```

```python
#Luodaan taas kuvaaja samalla tavalla kuten aiemmin ja katsotaan, mitä vaikutuksi tehdyillä muutoksilla on.
plt.figure(figsize=(16,10))
#Luodaan kuvaaja plot-komennolla kaikista Ruotsin tapauksista punaisella värillä ("r").
plt.plot(data_ruotsi.Date_reported, data_ruotsi.New_cases, color="r")
#Luodaan kuvaaja plot-komennolla kaikista Ruotsin koronaviruskuolemista mustalla värillä ("k").
plt.plot(data_ruotsi.Date_reported, data_ruotsi.New_deaths, color="k")
```

```python
#Huomataan, että vielä on nollarivit jäljellä (vasemmalla vaakasuorat viivat). Poistetaan vielä ne.
#Valitaan datasta käsittelyyn vain päivät 27.2.2020 jälkeen. 
data_ruotsi = data_ruotsi.loc[data_ruotsi['Date_reported']>='2020-02-27']
#Tulostetaan data
data_ruotsi
```

Näiden muokkauksien jälkeen luodaan taas kuvaaja vastaavalla tavalla kuin aiemmin:

```python
#Luodaan kuvaajalle valmis alue figure-komennolla ja määritellään sille koko.
plt.figure(figsize=(16,10))
#Luodaan kuvaaja plot-komennolla kaikista Ruotsin tapauksista punaisella värillä ("r").
cases, = plt.plot(data_ruotsi.Date_reported, data_ruotsi.New_cases, color="r")
#Luodaan kuvaaja plot-komennolla kaikista Ruotsin koronaviruskuolemista mustalla värillä ("k").
deaths, = plt.plot(data_ruotsi.Date_reported, data_ruotsi.New_deaths, color="k")
#Otsikoidaan kuva ja nimetään x- ja y-akselit.
plt.gca().set(title="COVID-19 tapaukset ja kuolemat Ruotsissa", xlabel="Päivämäärä",ylabel="Lukumäärä")
#Koska kuvaajia on useampi kuin yksi, lisätään niille oikeaan yläkulmaan selitteet(legend). Näin kuvan lukija tietää, mikä käyrä esittää mitäkin.
plt.legend((cases,deaths),("Tapaukset","Kuolemat"))
```

Jatketaan vielä kuvaajan kehittämistä. Voit ehkä huomata, että kuvaajat ovat "kohisevia", eikä niin selkeitä ja luettavia. Riippuu täysin kuvaajan käyttötarkoituksesta, voiko käyrää lähteä yksinkertaistamaan. Tehdään tässä tapauksessa molemmille kuvaajille hieman yleistävämpi esitys eli lasketaan liukuvasti keskiarvo tautitapauksille ja -uhreille muutama viikko.

```python
#Luodaan uusi muuttuja keskiarvodatalle ja syötetään sen arvoksi aluksi Ruotsin tautidata.
data_ruotsi_keskiarvo = data_ruotsi
#Lasketaan tapausten ja kuolemien liukuva keskiarvo rolling-komennolla. Asetetaan window-arvoksi 7 eli keskiarvo lasketaan 7 päivästä kerrallaan. Näin saadaan viikottainen data.
data_ruotsi_keskiarvo['New_cases']=data_ruotsi.New_cases.rolling(window=7).mean()
data_ruotsi_keskiarvo['New_deaths']=data_ruotsi.New_deaths.rolling(window=7).mean()

#Luodaan kuvaajalle valmis alue figure-komennolla ja määritellään sille koko.
plt.figure(figsize=(16,10))
#Luodaan kuvaaja plot-komennolla kaikista Ruotsin tapauksista punaisella värillä ("r").
mean_cases, = plt.plot(data_ruotsi_keskiarvo.Date_reported, data_ruotsi_keskiarvo.New_cases, color="r")
#Luodaan kuvaaja plot-komennolla kaikista Ruotsin koronaviruskuolemista mustalla värillä ("k").
mean_deaths, = plt.plot(data_ruotsi_keskiarvo.Date_reported, data_ruotsi_keskiarvo.New_deaths, color="k")
#Otsikoidaan kuva ja nimetään x- ja y-akselit.
plt.gca().set(title="COVID-19 tapaukset ja kuolemat Ruotsissa", xlabel="Päivämäärä",ylabel="Lukumäärä")
#Koska kuvaajia on useampi kuin yksi, lisätään niille oikeaan yläkulmaan selitteet(legend). Näin kuvan lukija tietää, mikä käyrä esittää mitäkin.
plt.legend((mean_cases,mean_deaths),("Tapaukset","Kuolemat"))

#Kokeile muuttaa window-arvoa. Mitä muutoksia huomaat kuvaajissa eri window-arvoilla?
```
